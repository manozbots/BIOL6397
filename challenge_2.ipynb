```python
import numpy as np
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def birthdeath(t,r,N):
    b = r*(t - (t**2)/N)
    d = r*(t**2)/N
    return b, d

####### define increse or decrease in population:
def increase(b,d):
    x = np.random.uniform(0,1,1)
    if x < b:
        a=1
    elif x < (b+d):
        a=-1
    else:
        a=0
    return a
```

```python
r = 0.004
K = 50
t = 3000
kk=100
p =100
N = 2*K

population=np.zeros([t,kk])
for j in range(kk):
    population[0,j]=p
    for k in range(t-1):
        if population[k,j] < 1:
            break
        else:
            b,d = birthdeath(population[k,j],r,N)
            population[k+1,j] = population[k,j] + increase(b,d)
### np.savetxt("population.csv", population, delimiter=",",header=' ')
```

```python
f = plt.figure(figsize=(3,3))
for i in range(kk):
    plt.plot(population[:,i],color= "C0",alpha=0.4)
plt.xlabel('time t')
plt.ylabel('population count')
```
